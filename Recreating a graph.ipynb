```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/DataSci Week 2 Underst Data/mortgages.csv'
dt = pd.read_csv(path)
dt
```

```python
mortgagefilt = dt['Mortgage Name'] == '30 Year'
interestfilt3 =  dt['Interest Rate'] == 0.03
interestfilt5 = dt['Interest Rate'] == 0.05
```

```python
mortgagefilt5 = dt.loc[mortgagefilt & interestfilt5]
mortgagefilt3 = dt.loc[mortgagefilt & interestfilt3]
```

```python
fig, ax = plt.subplots()
plt.style.use('seaborn')
ax.plot(mortgagefilt3['Month'], mortgagefilt3['Interest Paid'].cumsum(), c = 'g', label = '3 % Interest')
ax.plot(mortgagefilt5['Month'], mortgagefilt5['Interest Paid'].cumsum(), c = 'r', label = '5 % Interest')
plt.title('Cumulative Interest Paid Over 30 Year Loan', fontsize = 20)
plt.xlabel('Month', fontsize = 17)
plt.ylabel('Dollar Amount', fontsize = 17)
plt.xticks(fontsize = 13) and plt.yticks(fontsize = 13)
plt.legend(loc = (0.7,0.1));
plt.show()
```
