#2)  Transform the 'datetime' column into a datetime type and use it to create 3 new columns in the data frame containing the:

 1. Name of the Month
 2. Name of the Day of the Week
 3. Hour of the Day

* Make sure all 3 new columns are 'object' datatype so they can be one-hot encoded later.

* Drop the 'datetime' and 'season' columns. These are now redundant.

```python
df.info()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df.info()
```

* Name of the Month
* Name of the Day of the Week
* Hour of the Day

```python
df['Month - Name'] = df['datetime'].dt.month_name()
df['Day of the Week - Name'] = df['datetime'].dt.day_name()
df['Hour of the Day'] = df['datetime'].dt.hour
df.head()
```

* Make sure all 3 new columns are 'object' datatype so they can be one-hot encoded later.

```python
df.info()
```

Hour of the Day dtype is NOT an object, let's change it:

```python
df['Hour of the Day'] = df['Hour of the Day'].astype('object')
df.info()
```

* Drop the 'datetime' and 'season' columns. These are now redundant.

```python
df.drop(columns=['datetime', 'season'], inplace=True)
df.head()
```

# 3) The temperatures in the 'temp' and 'atemp' column are in Celsius. Use .apply() to convert them to Fahrenheit.

```python
# Taken from https://www.tutorialspoint.com/write-a-program-in-python-pandas-to-convert-a-dataframe-celsius-data-column-into-fahrenheit

df['temp'] = df.apply(lambda x: (9/5)*x['temp']+32, axis=1)
df['atemp'] = df.apply(lambda x: (9/5)*x['atemp']+32, axis=1)
df.head()
```

#4) Create a new column, 'temp_variance' that is the difference between 'temp' and 'atemp'. Drop the 'atemp' column.

```python
df['temp_variance'] = df['atemp'] - df['temp']
```

```python
df.drop(columns=['atemp'], inplace=True)
df.head()
```

**ENGINEERED FEATURES**

```python
df.isna().any().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
X = df.drop(columns='count')
y = df['count']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
```

```python
# Since this bulletpoint is optional,
# I Will continue to work on it after finishing Project 2 part 4.


# Since the result for the model prediction can be any* amount of hours,
# this would be a Regression problem. I decided that I'll use a LinearRegression or maybe a Random forest.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

* Pipeline:

```python
from sklearn.compose import make_column_selector, make_column_transformer
```

```python
cat_selector = make_column_selector(dtype_include='object')
num_selector = make_column_selector(dtype_include='number')

ohe = OneHotEncoder()
scaler = StandardScaler()

ohe_tuple = (ohe, cat_selector)
scaler_tuple = (scaler, num_selector)

col_trans = make_column_transformer(scaler_tuple, ohe_tuple)
```

```python
lin_reg = LinearRegression()
```

```python
pipe1 = make_pipeline(col_trans, lin_reg)
pipe1.fit(X_train, y_train);
```

* SCORES FOR ENGINEERED FEATURES:

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def eval_model(true, pred):
  """Takes true and predicted values (arrays) and prints MAE, MSE, RMSE and R2"""
  mae = mean_absolute_error(true, pred)
  mse = mean_squared_error(true, pred)
  rmse = np.sqrt(mse)
  r2 = r2_score(true, pred)

  print(f'MAE {mae},\n MSE {mse},\n RMSE: {rmse},\n R^2: {r2} ')

print('Train Evaluation')
eval_model(y_train, pipe1.predict(X_train))

print('\nTest Evaluation')
eval_model(y_test, pipe1.predict(X_test))
```
